```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_traj(fname):
    r = []
    q = []
    with open(fname, 'r') as f:
        inp_content = f.readlines()
        for i in inp_content:
            temp = i.split()
            r.append(float(temp[1]))
            q.append(float(temp[2]))
    return [r,q]
```

```python
rsteps = 71
qsteps = 45
for i in range(rsteps):
    plt.figure(figsize=[20,10])
    for j in range(qsteps):
        f_name = '/home/armin/vwall/usample_fib/1d_case/output/out.%d-%d.traj' % (j,i)
        [r, q] = read_traj(f_name)
        plt.xlim((-1,1))
        plt.hist(q, bins=50, density=True)
    plt.show()
```
